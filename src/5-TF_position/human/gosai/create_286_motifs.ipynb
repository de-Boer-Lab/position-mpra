```python
# Define the list of unique transcription factors
unique_tfs = [
    "JUNB", "NFE2", "KLF5", "SMARCC1", "FLI1", "GATA1", "EGR1", "PURA1", "SP3", "MAZ",
    "ZNF263", "ZNF281", "ARID3A", "FOXC1", "FOXP1", "FOXJ3", "FOXO1", "FUBP1", "IRF1",
    "IRF5", "GATA2", "TAL1", "RUNX1", "MYB", "KLF1", "STAT5A", "HOXB9", "CTCFL", "PITX1",
    "IKZF1", "EPAS1", "IRS2"
]

pattern = "|".join(unique_tfs)

filtered_df = motifs[motifs["Motif"].str.contains(pattern, case=False, na=False)]

print(filtered_df)
```
