```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the file path
file_path = "/scratch/st-cdeboer-1/sambina/position_mpra/data/TF_position/yeast/41587_2019_315_MOESM6_ESM.xlsx"

# Read the "positional" sheet
df = pd.read_excel(file_path, sheet_name="positional params - positional", engine="openpyxl", skiprows=1)
```

```python
df.head(5)
```

```python
df_ABF1 = df[df["TFAlias"] == "ABF1"]

# Sort by position for continuity
df_ABF1 = df_ABF1.sort_values(by="position")

# Create a figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(12, 12), sharex=True)  # 2 rows, 1 column

# Plot ABF1 (+ strand)
for dataset, group in df_ABF1.groupby("Dataset"):
    axes[0].plot(group["position"], group["positionalActivity"], linestyle='-', label=f"{dataset}")

axes[0].set_title("ABF1 (+ strand)")
axes[0].set_ylabel("Positional Activity")
axes[0].legend(title="Dataset")
axes[0].grid(True)

# Plot ABF1 (- strand)
for dataset, group in df_ABF1.groupby("Dataset"):
    axes[1].plot(group["position"], group["positionalActivityRC"], linestyle='-', label=f"{dataset}")

axes[1].set_title("ABF1 (- strand)")
axes[1].set_xlabel("Position")
axes[1].set_ylabel("Positional Activity")
axes[1].legend(title="Dataset")
axes[1].grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
