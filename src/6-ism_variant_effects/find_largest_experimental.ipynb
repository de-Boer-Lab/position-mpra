```python
data = pd.read_csv("/scratch/st-cdeboer-1/sambina/position_mpra/outputs/GTeX_experiment/42_variants.csv")



# assume your dataframe is called df with columns: base_id, centre, left, right

def pairwise_squared_diff_sum(row):
    return (row["centre"] - row["left"])**2 + (row["centre"] - row["right"])**2

# compute for each row
data["diff_sum"] = data.apply(pairwise_squared_diff_sum, axis=1)

# get the row with maximum value
max_row = data.loc[data["diff_sum"].idxmax()]

print("Row with largest pairwise squared diff sum:")
print(max_row)

```
