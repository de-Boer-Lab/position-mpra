```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/scratch/st-cdeboer-1/sambina/position_mpra/outputs/GTeX_experiment/42_variants.csv")
```

```python
heatmap_data = df[['left', 'centre', 'right']]

# Plot the heatmap
plt.figure(figsize=(6, len(df)))  # Adjust size based on number of rows
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, xticklabels=['left', 'centre', 'right'], yticklabels=df['base_id'])

plt.title("Heatmap of Left, Centre, Right")
plt.xlabel("Position")
plt.ylabel("variant")
plt.tight_layout()
plt.show()
```

```python
heatmap_data = df[['left', 'centre', 'right']]

# Set base_id as index for clustering labels
heatmap_data.index = df['base_id']

# Plot clustered heatmap
g = sns.clustermap(
    heatmap_data,
    cmap='coolwarm',       # red to blue
    metric='correlation',
    method='average',
    figsize=(6, len(heatmap_data)/2),
    annot=True,
    z_score=None,          # no normalization
    standard_scale=None,
    col_cluster=False
)
g.ax_heatmap.set_xlabel("Position")
g.ax_heatmap.set_ylabel("variant")

g.fig.suptitle("Clustered GTeX variants", y=1.05, fontsize=14, ha='center')

plt.show()
```

```python
heatmap_data = df[['left', 'centre', 'right']]

# Set base_id as index for clustering labels
heatmap_data.index = df['base_id']

# Plot clustered heatmap
g = sns.clustermap(
    heatmap_data,
    cmap='coolwarm',       # red to blue
    metric='euclidean',
    method='average',
    figsize=(6, len(heatmap_data)/2),
    annot=True,
    z_score=None,          # no normalization
    standard_scale=None,
    col_cluster=False
)
g.ax_heatmap.set_xlabel("Position")
g.ax_heatmap.set_ylabel("variant")

g.fig.suptitle("Clustered GTeX variants", y=1.05, fontsize=14, ha='center')

plt.show()
```
